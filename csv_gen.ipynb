```python
from ucimlrepo import fetch_ucirepo
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
import pandas as pd
```

```python
# Load the dataset
dataset = fetch_ucirepo(id=544)

# Data (as pandas dataframes)
X = dataset.data.features
y = dataset.data.targets['NObeyesdad'] # target
```

```python
# Identify categorical columns
categorical_cols = X.select_dtypes(include=['object']).columns

# Preprocessor for categorical variables
preprocessor = ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(), categorical_cols)
    ],
    remainder='passthrough'
)

# Fit and transform the data
X_transformed = preprocessor.fit_transform(X)
```

```python
# Get feature names without the onehot prefix
onehot_encoder = preprocessor.named_transformers_['onehot']
onehot_feature_names = onehot_encoder.get_feature_names_out(categorical_cols)

# Combine with the names of the remaining columns
remaining_columns = X.columns.difference(categorical_cols)
all_feature_names = list(onehot_feature_names) + list(remaining_columns)

# Convert the transformed data back to a DataFrame
X_transformed_df = pd.DataFrame(X_transformed, columns=all_feature_names)

# Concatenate the transformed features with the target column
final_df = pd.concat([X_transformed_df, y.reset_index(drop=True)], axis=1)
```

```python

# Save the DataFrame to a CSV file
final_df.to_csv('transformed_data.csv', index=False)

print("Data saved to 'transformed_data.csv'")
```
